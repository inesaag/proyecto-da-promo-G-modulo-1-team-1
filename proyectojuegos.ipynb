```python
from random import choices
import random
import unidecode
```

```python
class Juegos:
    def __init__(self) -> None:
        pass
    def piedrapapeltijera(self):
        lista_juego = ['piedra', 'papel', 'tijera']
        ronda=1
        victorias_usuario=0
        victorias_ordenador=0
        print(f'✂️🧻🪨 BIENVENIDO A PIEDRA, PAPEL O TIJERA 🪨🧻✂️')
        print('-----')
        while ronda <=3:
            pregunta = input('Piedra, papel o tijera').lower()
            eleccion = random.choice(lista_juego) 
            if pregunta == eleccion:
                print(f'{pregunta.capitalize()} es igual a {eleccion}')
                print('----')
    
            elif pregunta == "piedra" and eleccion == "tijera":
                print(f'RONDA {ronda}')
                print("¡Has ganado! 🪨 gana ✂️")
                victorias_usuario +=1
                ronda +=1
                print('----')
            elif pregunta == 'papel' and eleccion == 'piedra':
                print(f'RONDA {ronda}')
                print ("¡Has ganado! 🧻 gana 🪨")
                victorias_usuario +=1
                ronda +=1
                print('----')
            elif pregunta == 'tijera' and eleccion == 'papel':
                print(f'RONDA {ronda}')
                print("¡Has ganado! ✂️ gana 🧻")
                ronda +=1
                victorias_usuario +=1
                print('----')
            else:
                if eleccion == "piedra" and pregunta == "tijera":
                    print(f'RONDA {ronda}')
                    print("¡Has perdido! 🪨 gana ✂️")
                    victorias_ordenador +=1
                    ronda +=1
                    print('----')
                elif eleccion == 'papel' and pregunta == 'piedra':
                    print(f'RONDA {ronda}')
                    print("¡Has perdido! 🧻 gana 🪨")
                    victorias_ordenador +=1
                    ronda +=1
                    print('----')
                elif eleccion == 'tijera' and pregunta == 'papel':
                    print(f'RONDA {ronda}')
                    print("¡Has perdido! ✂️ gana 🧻")
                    victorias_ordenador +=1
                    ronda +=1
                    print('----')
        if victorias_usuario > victorias_ordenador:
            print(f'🎉🥳 ¡MUY BIEN!¡HAS GANADO! 🥳🎉')
        else:
            print(f'¡OH NO! ¡HAS PERDIDO! 😭')

    def preguntas(self):
        preguntas = {
            "¿Cuál es la capital de España?": "Madrid",
            "¿En qué continente se encuentra Egipto?": "África",
            "¿Cuál es el río más largo de Europa?": "Volga",
            "¿Cuál es la montaña más alta del mundo?": "Everest",
            "¿En qué país se encuentra el desierto del Sahara?": "Marruecos",
            "¿Cuál es la capital de Canadá?": "Ottawa",
            "¿Cuántos océanos hay en la Tierra?": "5",
            "¿Cuál es la capital de Australia?": "Camberra",
            "¿Cuál es el país más grande del mundo por área terrestre?": "Rusia",
            "¿En qué continente se encuentra la Gran Barrera de Coral?": "Oceanía",
            "¿Cuál es la capital de Argentina?": "Buenos Aires",
            "¿Cuántos estados tiene Estados Unidos?": "50",
            "¿En qué país se encuentra la ciudad de Petra?": "Jordania",
            "¿Cuál es la capital de Sudáfrica?": "Ciudad del Cabo",
            "¿En qué país se encuentra el volcán Kilimanjaro?": "Tanzania",
            "¿Cuál es el estrecho que separa Europa de África?": "Gibraltar",
            "¿En qué país se encuentra la ciudad de Kioto?": "Japón",
            "¿Cuál es el país más pequeño del mundo?": "Vaticano",
            "¿En qué continente se encuentra el río Amazonas?": "América",
            "¿Cuál es la capital de Turquía?": "Ankara",
            "¿En qué país se encuentra la Gran Muralla China?": "China",
            "¿Cuál es el punto más bajo de la Tierra?": "Mar Muerto",
            "¿En qué país se encuentra el Parque Nacional de Yellowstone?": "Estados Unidos",
            "¿En qué país se encuentra el monte Kilimanjaro?": "Tanzania",
            "¿En qué país se encuentra el Canal de Panamá?": "Panamá",
            "¿Cuál es la capital de Italia?": "Roma",
            "¿En qué país se encuentra la ciudad de Casablanca?": "Marruecos",
            "¿Cuál es la capital de India?": "Nueva Delhi",
            "¿En qué país se encuentra la ciudad de Cuzco?": "Perú",
            "¿En qué país se encuentra la isla de Bali?": "Indonesia",
            "¿Cuál es la capital de Rusia?": "Moscú",
            "¿En qué continente se encuentra el río Danubio?": "Europa",
            "¿Cuál es la capital de México?": "Ciudad de México",
            "¿En qué país se encuentra el desierto de Atacama?": "Chile",
            "¿En qué país se encuentra la ciudad de Dublín?": "Irlanda",
            "¿Cuál es la capital de Sudán?": "Jartum",
            "¿En qué continente se encuentra el río Yangtsé?": "Asia",
            "¿Cuál es la capital de Sudán del Sur?": "Yuba",
            "¿En qué país se encuentra la ciudad de Estambul?": "Turquía",
            "¿En qué continente se encuentra el río Támesis?": "Europa",
            "¿En qué país se encuentra el monte Fuji?": "Japón",
            "¿En qué continente se encuentra el río Misisipi?": "América",
            "¿Cuál es la capital de Colombia?": "Bogotá",
            "¿En qué país se encuentra el desierto del Gobi?": "Mongolia",
            "¿Cuál es la capital de Noruega?": "Oslo",
            "¿En qué continente se encuentra el río Congo?": "África",
        }
        puntos = 0
        ronda = 1
        print(f'🌍 BIENVENIDO A PREGUNTAS SOBRE GEOGRAFÍA 🌍')
        print('-----')
        while ronda <=10:

            pregunta, respuesta = choices(list(preguntas.items()), k=1)[0]
            preguntas.pop(pregunta)


            respuesta_minuscula = respuesta.lower()
            respuesta_notilde = unidecode.unidecode(respuesta_minuscula)

            respuesta_usuario = (input(pregunta))
            respuesta_usuario_minuscula = respuesta_usuario.lower()
            respuesta_usuario_notilde = unidecode.unidecode(respuesta_usuario_minuscula)

            if respuesta_usuario_notilde == respuesta_notilde:
                print(f"¡Muy bien! La respuesta correcta es {respuesta} 😎")
                puntos += 1
                ronda += 1
                print("---------------------")

            else:
                print("¡Oh no! Respuesta incorrecta, la respuesta correcta no era", respuesta_usuario, "sino", respuesta, '😉')
                ronda += 1
                print ('---------------------')

        if puntos == 0:
            print(f"No has acertado ninguna pregunta, deberías mirar algún mapa 🗺️🔍")

        elif 0< puntos < 4:
            print(f"Solo has acertado {puntos} preguntas, sigue practicando 🥲")
            
        elif 4 <= puntos <=6:
            print(f"Has acertado {puntos} preguntas, no está mal pero se puede mejorar 👍")

        elif 7<= puntos <=9:
            print(f"Has acertado {puntos} preguntas. ¡La geografia se te da genial! 💃")

        else: 
            print(f"¡Has acertado todas las preguntas! ¡Eres increíble! 🥳🎉")

    def ahorcado(self):
        animales = ["Elefante", "Tigre", "Jirafa", "Cebra", "Hipopótamo", "Oso", "Mono", "Lobo", "Gorila",
            "Pinguino", "Delfín", "Ballena", "Águila", "Leopardo", "Rinoceronte", "Halcón", "Jaguar", "Ratón", "León", "Orangután",
            "Camello", "Koala", "Panda", "Serpiente", "Tortuga", "Tucán", "Canguro", "Cocodrilo", "Hormiga"]
        palabra_secreta = random.choice(animales).lower()
        palabra_notilde = unidecode.unidecode(palabra_secreta)
        palabra_secreta1=list(palabra_notilde)
        caracter=[]

        for letra in palabra_secreta:
            caracter.append("_")
        caracter
        intentos = 1

        print(f'🐘🦁 BIENVENIDO AL AHORCADO. ¡Descubre el animal oculto! 🦁🐘')
        print('-----')
        
        while intentos <10 and "_" in caracter:
        
            letra= input("Inserta una letra:").lower()
            while letra in palabra_secreta1: 
                i=palabra_secreta1.index(letra)
                caracter.pop(i)
                caracter.insert(i,letra)
                palabra_secreta1.pop(i)
                palabra_secreta1.insert(i,"*")
                print(f"¡Bien! La {letra} está en la palabra 😎")
                print(caracter)
                print("------------")

                if  "_" not in caracter :
                    print("🎉🥳 ¡¡¡HAS GANADO!!! 🥳🎉")
                    break
            if letra not in palabra_secreta:
                print(f"La {letra} no está en la palabra 🥲")
                intentos +=1
                print("------------")
        
        if intentos >= 10:
            print("¡Oh no! Has perdido 😭")
            print(f'La palabra secreta era {palabra_secreta}')
```
